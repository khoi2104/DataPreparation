# EX1

```python
import pandas as pd
```

```python
df1 = pd.read_csv('fb.csv', index_col=[0], parse_dates=[0])
df1['ticker'] = 'FB'
df2 = pd.read_csv('aapl.csv', index_col=[0], parse_dates=[0])
df2['ticker'] = 'AAPL'
df3 = pd.read_csv('amzn.csv', index_col=[0], parse_dates=[0])
df3['ticker'] = 'AMZN'
df4 = pd.read_csv('nflx.csv', index_col=[0], parse_dates=[0])
df4['ticker'] = 'NFLX'
df5 = pd.read_csv('goog.csv', index_col=[0], parse_dates=[0])
df5['ticker'] = 'GOOG'

dataframe = pd.concat([df1, df2, df3, df4, df5], axis=0, join='inner').sort_index()
dataframe
```

```python
df = dataframe
pd.DataFrame.to_csv(df, 'faang.csv')
```

```python
df.index = pd.to_datetime(df.index)
df.head(3)
```

```python
df.volume = df.volume.astype('int')
df.head(3)
```

```python
df.sort_values(by=['date', 'ticker'])
df.head(3)
```

```python
df.sort_values(
    by='volume', ascending=True
).head(7)
```

# EX2

```python
dataframe1 = pd.read_csv('covid19_cases.csv')
covid = dataframe1
covid.head(3)
 
```

```python
covid['date'] = pd.to_datetime(covid.dateRep)
covid.head(3)
```

```python
covid.set_index('date', inplace=True)
covid.sort_index().head(3)
```

```python
covid.countriesAndTerritories.replace(
    to_replace=['United_States_of_America', 'United_Kingdom'],
    value=['USA', 'UK']
)
```

```python
specific_countries = covid[covid.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
specific_countries
cases = specific_countries.groupby('countriesAndTerritories')['cases']
cases.sum()

# ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA']
```

```python
covid.fillna(0, inplace=True)

pivoted_covid =  covid.pivot(
    columns='countriesAndTerritories',
    values='cases'
)

pivoted_covid.head(3)
```
